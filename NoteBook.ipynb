Import des librairies et declarations des variables necessaires au bon exécution du programme


```python
# Import des librairy necessaire
import utils
import csv
import datetime
import sys

# Lecture des data
link = "data\\ventes.csv"

with open(link, mode="r", encoding="utf-8") as file:
    data = file.read().splitlines()

# Déclaration des variables
data = csv.reader(data)
en_tete = next(data)
data = list(data)
date = datetime.datetime.now()
```

#### Définitions et test des differentes fonctions


Fonction permettant d'afficher toutes les ventes enregistrées


```python
# def Show_all(data):
for line in data:
    print(line)
```

La fonction suivante, comme son nom l'indique, elle renvoie le chiffre d'affaire total


```python
# def Chiffre_affaire_total(data):
chiffre_affaire_total = 0
for line in data:
    if float(line[3]) > 0 and float(line[4]) > 0:
        chiffre_affaire = float(line[3]) * float(line[4])
        chiffre_affaire_total += chiffre_affaire
    else:
        chiffre_affaire_total += 0

print(f"Chiffre d'affaire total $ {round(chiffre_affaire_total, 3)}")
# return round(chiffre_affaire_total, 3)
```

Celle ci permet d'affichier les nombres de ventes par villes


```python
# def Ventes_par_ville(data):
ventes_par_ville = {}

for line in data:
    if line[-1] in ventes_par_ville:
        ventes_par_ville[line[-1]] += 1
    else:
        ventes_par_ville[line[-1]] = 1
print(f"{ventes_par_ville}")
```

Celle ci permet d'affichier les nombres de ventes par categories


```python
# def Ventes_par_categorie(data):
ventes_par_categorie = {}

for line in data:
    if line[2] in ventes_par_categorie:
        ventes_par_categorie[line[2]] += 1
    else:
        ventes_par_categorie[line[2]] = 1
print(f"{ventes_par_categorie}")
```

Celle ci renvoie la categorie ayant enrégistré le plus de ventes


```python
# def Best_category(data):
categories = {}
best_category = {}
for line in data:
    if line[2] in categories:
        categories[line[2]] += 1
    else:
        categories[line[2]] = 1

for category, sales in categories.items():
    if sales == max(categories.values()):
        best_category = {category: sales}

print(best_category)
# return best_category
```
